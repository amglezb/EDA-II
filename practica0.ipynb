# Algoritmo de Complejidad: Constante

```python
import matplotlib.pyplot as plt

arr = list(range(100))
time = 0

def search(arr, val, time):
    for h in range(0, 100):
        time += 1
        if arr[h] == val:
            return time

x = []
y = []

for i in range(0, len(arr)):
    x.append(i)
    y.append(search(arr, arr[99], time))

plt.plot(x, y)
plt.show()
```
