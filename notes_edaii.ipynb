## **Búsqueda por expansión (amplitud, anchura) (BFS) - Breadth-First Search**
- $G = {V, A}$
- Parte de un nodo fuente
    - Lo enqueued
- Usamos una queue
    - Nodo Borned -> White
    - Nodo Discovered -> Gray
    - Nodo Visited -> Black
- Se Recorren los nodos del nodo fuente y se pintan en gris estos son Nodos Discovered
- Cuando ya se recorrieron los nodos del nodo fuente este es un nodo Visited

- Al final se obtiene un grafo en forma de arbol -> estado final del grafo o arbol (dependiendo del nodo fuente el arbol cambia)

- El orden para iniciar la búsqueda lo da la lista de adyacencia

- Grafos con nodos no conectados, BFS crea un bosque

- Complejidad de la Creación de la lista de adyacencia:
- n + m
- n = Vertices
- m = Aristas $= n(n-1)/2$
    - O(n^2)

- Complejidad temporal de la Búsqueda:
    - n = V
    - m = A
        - O(n+m)

```python
def hypercube(self, d): # ----- Hipercubo -----
    num_nodos_esperado = 2**d
    if self.numNodes != num_nodos_esperado:
        print("Error: El grafo debe inicializarse con " + str(num_nodos_esperado) + " nodos para un hipercubo de dimensión " + str(d) + ".")
        return
    
    for i in range(num_nodos_esperado):
        for j in range(d):
            vecino = i ^ (1 << j)
            if i < vecino:
                self.insert_edge(i + 1, vecino + 1, 1, self.directed)
```

Todos los subárboles son conjuntos disjuntos

Todas los nodos de la izquierda deben ser menor y a la derecha mayores

En una matriz:

Fila 1 -> Nodos
Fila 2 -> Predecesor

Un árbol binario es aquel que solo tiene un hijo izquierdo y un hijo derecho

Todas sus operaciones son en tiempo log_2(n)

Notaciones:
- Infíja: Jerarquía de Operaciones
- Prefija: Operadores antes de, Paréntesis invisible en pares $(*+ABC) -> (A+B) * C$
- Posfija: Operadores despues de cada operación, $(AB+C*) -> (A+B)*C$

Para los árboles hay 3 recorridos (infijo, prefijo y sufijo (posrtfijo))
- Recorrido Infijo (InOrden)
    - Recorrer subarbol izquierda
    - Visitamos raíz
    - Recorrer el subárbol derecho
- Recorrido Prefijo (PreOrden)
    - Visitamos raíz
    - Recorrer subarbol izquierda
    - Recorrer el subárbol derecho
- Recorrido PostOrden (PosOrden)
    - Recorrer subarbol izquierda
    - Recorrer el subárbol derecho
    - Visitamos raíz
