## **Bubble Sort**
### **Algoritmo:** (y polinomios)

```python
import random
import matplotlib.pyplot as plt

MAX = 100

def bubble_sort(C):
    n = len(C)                  # 4                 4
    i = 0                       # 3                 3
    while i < n:                # (n+1)*4           4n + 4
        j = 0                   # n*3               3n
        while j < n - i - 1:    # n*((n/2)+1)*7     (7n^2)/2 + 7n
            if C[j] > C[j + 1]: # n*(n/2)*7         (7n^2)/2            # Mejor Caso: no se entra al if solo se evalua -> (7n^2)/2
                temp = C[j]     # n*(n/2)*4         2n^2                # 0             # n*(n/4)*4     # n^2
                C[j] = C[j + 1] # n*(n/2)*6         3n^2                # 0             # n*(n/4)*6     # (3n^2)/2 
                C[j + 1] = temp # n*(n/2)*6         3n^2                # 0             # n*(n/4)*6     # (3n^2)/2 
            j = j + 1           # n*(n/2)*4         2n^2                                # Caso Promedio: Se realizan en promedio la mitad de las asignaciones
        i = i + 1               # n*4               4n
C = []                          #       Peor Caso = 17n^2 + 18n + 11
                                #                           Mejor Caso = 9n^2 + 18n + 11
                                #                                       Caso Promedio = 13n^2 + 18n + 11
for i in range(1, MAX + 1):
    C.append(random.randint(-3000, 3000))

print(C)
bubble_sort(C)
print(C)
```

### **Algoritmo** (Análisis)):

```python
def bubble_sort_graph(C):
    times = 0
    n = len(C)
    i = 0
    while i < n:
        j = 0
        swapped = False
        times += 1
        while j < n - i - 1:
            times += 1
            if C[j] > C[j + 1]:
                temp = C[j]
                C[j] = C[j + 1]
                C[j + 1] = temp
                swapped = True
            j = j + 1
        if swapped == False:
            break
        i = i + 1
    return times
```

### **Mejor Caso** (Lista Ordenada Ascendente):

```python
x = []
y_mejor = []
C_mejor = []

for n in range(1, MAX + 1): # checar la grafica
    C_mejor.append(n)
    x.append(len(C_mejor))
    y_mejor.append(bubble_sort_graph(C_mejor))

plt.plot(x, y_mejor)
plt.grid(True)
plt.show()
```

### **Peor Caso** (Lista Ordenada Descendente):

```python
x = []
y_peor = []
C_peor = []

for n in range(1, MAX + 1):
    C_peor.append(MAX - n)
    x.append(len(C_peor))
    y_peor.append(bubble_sort_graph(C_peor))

plt.plot(x, y_peor)
plt.grid(True)
plt.show()
```

### **Caso Promedio** (Lista de Valores aleatorios):

```python
x = []
y_prom = []
C_prom = []

for n in range(1, MAX + 1):
    C_prom.append(random.randint(-1000, 1000))
    x.append(len(C_prom))
    y_prom.append(bubble_sort_graph(C_prom))

plt.plot(x, y_prom)
plt.grid(True)
plt.show()
```

## **Merge Sort**
### **Algoritmo:** (y polinomios)

```python
MAX = 1000

def merge_sort(B):                          # Polinomio T(n) = 2T(n/2)+(55+14)n+19+38+3+4 = 2T(n/2)+69n+64
    n = len(B)                              # 4
    if n < 1 or n == 1:                     # 7
        return B                            # 2
    m = n // 2                              # 5
    L = []                                  # 3
    i = 0                                   # 3
    while i < m:                            # 4((n/2)+1)
        L.append(B[i])                      # 4(n/2)
        i = i + 1                           # 4(n/2)
    R = []                                  # 3
    j = m                                   # 3
    while j < n:                            # 4((n/2)+1)
        R.append(B[j])                      # 4(n/2)
        j = j + 1                           # 4(n/2)         todo lo anterior =14n+38
    L = merge_sort(L)                       # T(n/2)+2
    R = merge_sort(R)                       # T(n/2)+2
    return merge(L, R)                      # T_merge(n) + 3

def merge(L, R):                            # 19+25k+15nL+15nR = 19+25k+15n
                                            # Polinomio k = n/2        -> 55n + 19
    B_merge = []                            # 3
    i = 0                                   # 3
    j = 0                                   # 3
    nL = len(L)                             # 4
    nR = len(R)                             # 4
    while i < nL and j < nR:                # 8k -> k = (nL + nR - 1)
        iL = L[i]                           # 4k
        jR = R[j]                           # 4k
        if iL < jR or iL == jR:             # 6k
            B_merge.append(iL)              # 3k
            i = i + 1                       # 4k
        else:                               # 1k
            B_merge.append(jR)              # 3k
            j = j + 1                       # 4k
                                            # -> se simplifica a 4nL porque i recorre 0..nL
    while i < nL:                           # 4(nL-i+1)     -> 4nL
        iL = L[i]                           # 4(nL-i+1)     -> 4nL
        B_merge.append(iL)                  # 3(nL-i+1)     -> 3nL
        i = i + 1                           # 4(nL-i+1)     -> 4nL
                                            # -> se simplifica a 4nR porque j recorre 0..nR
    while j < nR:                           # 4(nR-j+1)     -> 4nR
        jR = R[j]                           # 4(nR-j)       -> 4nR
        B_merge.append(jR)                  # 3(nR-j)       -> 3nR
        j = j + 1                           # 4(nR-j)       -> 4nR
                                            # nR + nL = n
    return B_merge                          # 2
B = []
for i in range(MAX + 1):
    B.append(MAX - i)

print(B)
print(merge_sort(B))
```

### **Algoritmo** (Análisis)):

```python
def merge_sort_graph(B, times):
    sorted_list, times = merge_sort(B, times)
    return times

def merge_sort(B, times):
    if len(B) <= 1:
        return B, times
    m = len(B) // 2
    L = []
    i = 0
    while i < m:
        L.append(B[i])
        i = i + 1
        times = times + 1
    R = []
    j = m
    while j < len(B):
        R.append(B[j])
        j = j + 1
        times = times + 1
    L, times = merge_sort(L, times)
    R, times = merge_sort(R, times)
    B_merge, times = merge(L, R, times)
    return B_merge, times


def merge(L, R, times):
    B_merge = []
    i = j = 0
    while i < len(L) and j < len(R):
        times = times + 1
        if L[i] <= R[j]:
            B_merge.append(L[i])
            i = i + 1
        else:
            B_merge.append(R[j])
            j = j + 1
    while i < len(L):
        times = times + 1
        B_merge.append(L[i])
        i = i + 1
    while j < len(R):
        times = times + 1
        B_merge.append(R[j])
        j = j + 1
    return B_merge, times
```
