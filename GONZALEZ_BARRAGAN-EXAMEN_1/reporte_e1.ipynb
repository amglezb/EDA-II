# **Descripci√≥n de la Funci√≥n Hash**

Mi funci√≥n hash es de **mezcla de bits**.

El proceso se hace en 5 l√≠neas
1. Mezcla
    1. Se desplazan bits 13 posiciones a la derecha
    2. Se mezclan los bits originales con los desplazados
    3. Se multiplica por 1000000007 (n√∫mero primo) para dispersar m√°s los bits
2. Mezcla
    1. Se desplazan bits 14 posiciones a la derecha
    2. Se mezclan los bits originales con los desplazados
    3. Se multiplica por 2147483647 (n√∫mero primo) para dispersar m√°s los bits
3. Mezcla
    1. Se desplazan bits 15 posiciones a la derecha
    2. Se mezclan los bits originales con los desplazados
    3. Se multiplica por 5555555557 (n√∫mero primo) para dispersar m√°s los bits
4. Mezcla
    1. Se desplazan bits 16 posiciones a la derecha
    2. Se mezclan los bits originales con los desplazados
5. Normalizaci√≥n
    1. Para evitar √≠ndices negativos se obtiene el valor absoluto
    2. Se aplica m√≥dulo de n para mantener el √≠ndice entre 0 y (n-1)

# **Polinomio**
El algoritmo es cte. $O(1)$ por lo tanto los polinomios son:

Temporal: $39$

Espacial: $8$

En todos los casos. (Calculo en la siguiente secci√≥n)

# **C√≥digo**

```python
N = 1000

def myHash(value):                               # Temporal      Espacial
  value = (value ^ (value >> 13)) * 1000000007      # 9             # 2
  value = (value ^ (value >> 14)) * 2147483647      # 9             # 2
  value = (value ^ (value >> 15)) * 5555555557      # 9             # 2
  value = value ^ (value >> 16)                     # 7             # 2
  return abs(value) % N                             # 5
            # Polinomio del best/worst/average case = 39            = 8
```
